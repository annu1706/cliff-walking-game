```python
import numpy as np
import matplotlib.pyplot as plt
import gymnasium as gym
```

```python
sarsa_rewards = np.load("sarsa_rewards.npy")
q_rewards = np.load("q_rewards.npy")

print("SARSA Episodes :", len(sarsa_rewards))
print("Q-Learning Episodes :", len(q_rewards))
```

```python
plt.figure(figsize=(12,6))

plt.plot(sarsa_rewards, label="SARSA")
plt.plot(q_rewards, label="Q-Learning")

plt.title("Episode Reward Comparison")
plt.xlabel("Episode")
plt.ylabel("Reward")

plt.grid(True)
plt.legend()

plt.show()
```

```python
window = 20

sarsa_avg = np.convolve(
    sarsa_rewards,
    np.ones(window)/window,
    mode="valid"
)

q_avg = np.convolve(
    q_rewards,
    np.ones(window)/window,
    mode="valid"
)

plt.figure(figsize=(12,6))

plt.plot(sarsa_avg, label="SARSA")
plt.plot(q_avg, label="Q-Learning")

plt.title("Moving Average Reward")
plt.xlabel("Episode")
plt.ylabel("Average Reward")

plt.grid(True)
plt.legend()

plt.show()
```

```python
print("="*40)
print("SARSA")
print("="*40)

print("Average Reward :", np.mean(sarsa_rewards))
print("Maximum Reward :", np.max(sarsa_rewards))
print("Minimum Reward :", np.min(sarsa_rewards))

print()

print("="*40)
print("Q-LEARNING")
print("="*40)

print("Average Reward :", np.mean(q_rewards))
print("Maximum Reward :", np.max(q_rewards))
print("Minimum Reward :", np.min(q_rewards))
```

```python
print("="*50)

if np.mean(sarsa_rewards) > np.mean(q_rewards):
    print("🏆 SARSA performed better on average.")
elif np.mean(sarsa_rewards) < np.mean(q_rewards):
    print("🏆 Q-Learning performed better on average.")
else:
    print("🤝 Both algorithms performed similarly.")

print("="*50)
```
